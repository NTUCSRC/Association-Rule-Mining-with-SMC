# Apriori Algorithms

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv('../data/clean/diabetes_all.csv')
df.shape
```

```python
df_apriori = apriori(df, min_support=0.3, use_colnames=True)
```

```python
association_rules(df_apriori, metric="confidence", min_threshold=0.7)
```
