# Apriori Algorithms

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# ignore warning
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('../data/clean/diabetes_all.csv')
df.shape
```

```python
df_apriori = apriori(df, min_support=0.25, use_colnames=True)
df_apriori.to_csv('../data/clean/diabetes_apriori.csv', index=False)
df_apriori.head(20)
```

```python
df_association= association_rules(df_apriori, metric="confidence", min_threshold=0.6)
df_association.to_csv('../data/clean/association_rules.csv', index=False)

df_association.sort_values(by='confidence', ascending=False, inplace=True)
df_association.head(20)
```
